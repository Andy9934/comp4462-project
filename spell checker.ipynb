```python
#load dataset
import pandas as pd
df_text = pd.read_csv('YInt.csv')
df_text
```

```python
df_text.message = df_text.message.astype(str)
```

```python
#remove string start with @
import re
df_text_no_at = df_text
df_text_no_at['message'] = df_text_no_at['message'].str.replace('@(\w+)', '', regex=True)
df_text_no_at
```

```python
#remove @ name and split hashtag
import re
df_text_hashtag = df_text_no_at
df_text_hashtag['hashtag'] = df_text_hashtag['message'].str.findall('#(\w+)')
df_text_hashtag['message'] = df_text_hashtag['message'].str.replace('#(\w+)','', regex=True)
df_text_hashtag
```

```python
import pkg_resources
from symspellpy import SymSpell

sym_spell = SymSpell(max_dictionary_edit_distance=5, prefix_length=12)
dictionary_path = pkg_resources.resource_filename(
    "symspellpy", "frequency_dictionary_en_82_765.txt"
)
bigram_path = pkg_resources.resource_filename(
    "symspellpy", "frequency_bigramdictionary_en_243_342.txt"
)
# term_index is the column of the term and count_index is the
# column of the term frequency
sym_spell.load_dictionary(dictionary_path, term_index=0, count_index=1)
sym_spell.load_bigram_dictionary(bigram_path, term_index=0, count_index=2)

# lookup suggestions for multi-word input strings (supports compound
# splitting & merging)
input_term = df_text_hashtag['message'][0]
# max edit distance per lookup (per single word, not per whole input string)
suggestions = sym_spell.lookup_compound(input_term, max_edit_distance=5)
# display suggestion term, edit distance, and term frequency
for suggestion in suggestions:
    print(suggestion)
```

```python
df_text_hashtag['message'] = df_text_hashtag['message'].map(lambda x: sym_spell.lookup_compound(x, max_edit_distance=5))
df_text_hashtag
```

```python
df_new = df_text_hashtag
df_new['message'] = df_new['message'].str[0]
df_new
```

```python
df_new.to_csv('df_text.csv')
```
